```python
def splitColumn(current):
    if(type(current) == str):
        return current.split("**")
```

```python
data["Ingredients"] = data["Ingredients"].apply(splitColumn)
```

```python
data["Directions"] = data["Directions"].apply(splitColumn)
```

```python
def toDict(recipe):
    if(type(recipe) == list):
        myDict = {}
        i = -1
        for item in recipe:
            if type(item) == str and item not in myDict:
                item = item.replace("'", "")
                key = i + 1
                i += 1
                myDict[str(key)] = item
        return  ast.literal_eval(json.dumps(myDict))
```

```python
data["Ingredients"] = data["Ingredients"].apply(toDict)
data["Directions"] = data["Directions"].apply(toDict)
```

```python
test = data["Ingredients"][2]
allFoodNames = set(allFood["name"].tolist())
test
```

```python
def getRawFood(ingredientList):
    if(type(ingredientList) == dict):
        rawFood = set()
        for k in ingredientList.items():
            temp = set(k[1].split(" "))
            for food in allFoodNames.intersection(temp):
                if len(food) > 0: 
                    rawFood.add(food)
        rawFood = list(rawFood)
        return rawFood
```

```python
data["rawFood"] = data["Ingredients"].apply(getRawFood)
```
